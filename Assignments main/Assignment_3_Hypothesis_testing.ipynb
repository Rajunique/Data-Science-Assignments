```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#solution for cutlets problem
# Ha: there is a significance difference in the diameter of the cutlets
# H0: there is no significance difference in the diameter of the cutlets
```

```python
df = pd.read_csv("/content/Cutlets (1).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
unit_A=pd.Series(df.iloc[:,0])
unit_A
```

```python
unit_B=pd.Series(df.iloc[:,1])
unit_B
```

```python
n1=len(unit_A)
p1=0.5

n1
```

```python
cutlet1=np.random.binomial(1,p1,n1)
```

```python
cutlet1
```

```python
n2=len(unit_B)
p2=0.5

n2
```

```python
cutlet2=np.random.binomial(1,p2,n2)
```

```python
cutlet2
```

```python
stats.ttest_ind(cutlet1,cutlet2)
```

Significant value is 5% and pvalue is 63%
Since p value is greater than significance value

NULL HYPOTHESIS IS TRUE
# No difference in the diameter of the cutlet between the two units

```python
#solution for labTAT question
# Anova testing
#Ha:There is a difference in average TAT among the different laboratories
#H0:There is no difference in average TAT among the different laboratories
```

```python
df1=pd.read_csv("/content/LabTAT (1).csv")
```

```python
df1
```

```python
stats.f_oneway(df1["Laboratory 1"],df1["Laboratory 2"],df1["Laboratory 3"],df1["Laboratory 4"])
```

Significance = 0.05
P-VAlue=2.11 which is greater than 0.05


Hence,
NULL HYPOTHESIS ACCEPTED(H0)


There is no difference in average TAT among the different laboratories and reject the alternate hypothesis
