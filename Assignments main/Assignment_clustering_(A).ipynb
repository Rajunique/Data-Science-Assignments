# **K-MEANS CLUSTERING**

```python
df_k.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df_k)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.lineplot(x=range(1,11),y=wcss)
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmeans=KMeans(n_clusters=5)
predict=kmeans.fit_predict(df_k)
```

```python
predict
```

```python
df_k.head()
```

```python
df_ori=df_k.copy()
```

```python
df_ori['clusterid']=kmeans.labels_
```

```python
df_ori
```

```python
kmeans.cluster_centers_
```

```python
df_k1=df_ori.groupby("clusterid").mean().reset_index()
df_k1
```

```python
df_ori["Balance"][df_ori["clusterid"]==3]
```

```python
label_0=df_ori[predict==0]
label_1=df_ori[predict==1]
label_2=df_ori[predict==2]
label_3=df_ori[predict==3]
label_4=df_ori[predict==4]
```

```python
label_3
```

```python
cols=df_ori.columns

plt.scatter(label_0[cols[2]], label_0[cols[3]], color='red')
plt.scatter(label_1[cols[2]], label_1[cols[3]], color='green')
plt.scatter(label_2[cols[2]], label_2[cols[3]], color='blue')
plt.scatter(label_3[cols[2]], label_3[cols[3]], color='black')
plt.scatter(label_4[cols[2]], label_4[cols[3]], color='purple')
plt.show()
```

```python
df_ori.columns
```

```python
plt.scatter(label_0[cols[3]], label_0[cols[4]], color='red')
plt.scatter(label_1[cols[3]], label_1[cols[4]], color='green')
plt.scatter(label_2[cols[3]], label_2[cols[4]], color='blue')
plt.scatter(label_3[cols[3]], label_3[cols[4]], color='black')
plt.scatter(label_4[cols[3]], label_4[cols[4]], color='purple')
plt.show()
```

K_Means **Accuracy**

```python
silhouette_score(df_ori,df_ori["clusterid"])
```

# **DBSCAN CLUSTERING**

```python
df_b.head()
```

```python
from sklearn.neighbors import NearestNeighbors

nn=NearestNeighbors(n_neighbors=8)

nbrs=nn.fit(df_b)

distances,indices=nbrs.kneighbors(df_b)
```

```python
from sklearn.cluster import DBSCAN
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
distances
```

```python
df_b.columns
```

```python
dbscan=DBSCAN(eps=0.1,min_samples=13)
dbscan.fit_predict(df_b)
```

```python
dbscan.labels_
```

```python
df_b["clusterdb"]=dbscan.labels_
df_b.head()
```

```python
df_b[df_b["clusterdb"]==1]
```

# **DBSCAN accuracy**

```python
silhouette_score(df_b,df_b["clusterdb"])
```
